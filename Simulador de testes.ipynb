```python
quiz3=False
quiz4=True
quiz5=False

perguntasLink = r"perguntas e respostas\quiz_num\perguntas_com_opcoes.txt"
respostasLink = r"perguntas e respostas\quiz_num\respostas_corretas.txt"

if quiz3:
    perguntasLink = perguntasLink.replace("quiz_num", "quiz 3")
    respostasLink = respostasLink.replace("quiz_num", "quiz 3")
     
if quiz4:
    perguntasLink = perguntasLink.replace("quiz_num", "quiz 4")
    respostasLink = respostasLink.replace("quiz_num", "quiz 4")
    
if quiz5:
    perguntasLink = perguntasLink.replace("quiz_num", "quiz 5")
    respostasLink = respostasLink.replace("quiz_num", "quiz 5")
   

perguntas_com_opcoes = [
]

respostas_corretas = [
   ]

# Read questions and answers from files
with open(perguntasLink, "r", encoding="utf-8") as f:
    content = f.read()
    # Split by """, and remove empty strings
    questions_raw = [q + '""",' for q in content.split('""",') if q.strip()]
    # Clean up each question by removing the starting """ and trailing """,
    perguntas_com_opcoes = [q.strip().strip('"""').strip() for q in questions_raw]

# Remove empty strings and clean up questions
perguntas_com_opcoes = [question.strip() for question in perguntas_com_opcoes]
perguntas_com_opcoes = [question.replace("**", "") for question in perguntas_com_opcoes]



with open(respostasLink, "r", encoding="utf-8") as f:
    content = f.read()
    # Add line breaks after commas
    content = content.replace(', "', ',"\n')
    content = content.replace("\\", "")
    # Split by "], and remove empty strings
    answers_raw = [a + "]," for a in content.split("],") if a.strip()]
    # Clean up each answer set by removing the starting [ and trailing ],
    answers_cleaned = [a.strip().strip('[]').strip() for a in answers_raw]
    # Split individual answers by comma and clean them
    respostas_corretas = [
        [ans.strip().strip('"').strip() for ans in answer.split('","')]
        for answer in answers_cleaned
    ]



from IPython.display import clear_output, display, Markdown

perguntas_com_opcoes = [question.replace("**", "") for question in perguntas_com_opcoes]

import tkinter as tk
from tkinter import ttk
from tkinter import messagebox

class QuizInterface:
    def __init__(self, root):
        self.root = root
        self.root.title("Technical Debt Quiz")
        self.root.geometry("800x600")
        
        # Combine questions and answers
        self.all_questions = perguntas_com_opcoes 
        self.all_answers = respostas_corretas 
        self.current_question = 0
        
        # Create main frame
        self.main_frame = ttk.Frame(root, padding="10")
        self.main_frame.grid(row=0, column=0, sticky=(tk.W, tk.E, tk.N, tk.S))
        
        # Question number label
        self.question_num = ttk.Label(self.main_frame, text="Question 1/12")
        self.question_num.grid(row=0, column=0, pady=10)
        
        # Question text
        self.question_text = tk.Text(self.main_frame, wrap=tk.WORD, height=15, width=80)
        self.question_text.grid(row=1, column=0, pady=10)
        
        # Answer text
        self.answer_text = tk.Text(self.main_frame, wrap=tk.WORD, height=10, width=80)
        self.answer_text.grid(row=2, column=0, pady=10)
        
        # Buttons
        self.button_frame = ttk.Frame(self.main_frame)
        self.button_frame.grid(row=3, column=0, pady=10)
        
        self.show_answer_btn = ttk.Button(self.button_frame, text="Show Answer", command=self.show_answer)
        self.show_answer_btn.grid(row=0, column=0, padx=5)
        
        self.next_btn = ttk.Button(self.button_frame, text="Next Question", command=self.next_question)
        self.next_btn.grid(row=0, column=1, padx=5)
        
        self.prev_btn = ttk.Button(self.button_frame, text="Previous Question", command=self.prev_question)
        self.prev_btn.grid(row=0, column=2, padx=5)
        
        # Show first question
        self.show_question()
        
    def show_question(self):
        self.question_num.config(text=f"Question {self.current_question + 1}/{len(self.all_questions)}")
        self.question_text.delete(1.0, tk.END)
        self.question_text.insert(tk.END, self.all_questions[self.current_question])
        self.answer_text.delete(1.0, tk.END)
        
    def show_answer(self):
        self.answer_text.delete(1.0, tk.END)
        answers = self.all_answers[self.current_question]
        for answer in answers:
            self.answer_text.insert(tk.END, f"â€¢ {answer}\n")
            
    def next_question(self):
        if self.current_question < len(self.all_questions) - 1:
            self.current_question += 1
            self.show_question()
        else:
            messagebox.showinfo("End", "You've reached the end of the quiz!")
            
    def prev_question(self):
        if self.current_question > 0:
            self.current_question -= 1
            self.show_question()

# Create and run the interface
root = tk.Tk()
app = QuizInterface(root)
root.mainloop()
```
